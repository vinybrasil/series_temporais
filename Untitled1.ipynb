```python
t = np.linspace(1, 24, 24)
```

```python
t
```

```python
x = np.array([t])
```

```python
x = x.reshape(1, 24, 1)
```

```python
y = np.array([data[0:24].Sales.values])
```

```python
y
```

```python
model = Sequential()
model.add(LSTM(24, input_shape=(24,1)))
model.add(Dense(8))
model.add(Dense(24))
model.compile(loss='mae', optimizer='adam')
history = model.fit(x, y, epochs = 350, batch_size=32, verbose=1, shuffle=False)
plt.plot(history.history['loss'])
```

```python
model.predict(np.array([t]).reshape(1, 24,1))
```
