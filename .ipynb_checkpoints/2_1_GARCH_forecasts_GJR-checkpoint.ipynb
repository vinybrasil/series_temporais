```python
#source: https://arch.readthedocs.io/en/latest/univariate/univariate_volatility_modeling.html

import warnings
warnings.simplefilter('ignore')

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn

seaborn.set_style('darkgrid')
plt.rc("figure", figsize=(16, 6))
plt.rc("savefig", dpi=90)
plt.rc("font",family="sans-serif")
plt.rc("font",size=14)

import datetime as dt

import arch.data.sp500
```

```python
#carregando os dados

st = dt.datetime(1988, 1, 1)
en = dt.datetime(2018, 1, 1)
data = arch.data.sp500.load()
market = data['Adj Close']
returns = 100 * market.pct_change().dropna()
ax = returns.plot()
xlim = ax.set_xlim(returns.index.min(), returns.index.max())
```

<h1>GARCH com mdia constante</h1>
$$ r_{t} = \mu + \epsilon_{t}$$
$$\epsilon_{t} = \sigma_{t} e_{t}$$
$$\sigma^{2}_{t} = \omega + \alpha \epsilon^{2}_{t-1} + \beta \sigma^{2}_{t-1}$$
$$\epsilon_{t}  \sim N(0,1)$$

```python
#mdia constante, GARCH(1,1) para a variancia condicial e erro normais

from arch import arch_model

am = arch_model(returns)
res = am.fit(update_freq=5)
```

```python
res
```

```python
fig = res.plot(annualize='D')
```

```python
#separando at essa data e o resto ele faz o forecast
split_date = dt.datetime(2010,1,1) 
res = am.fit(last_obs=split_date)
```

### Analytical forecast (tem a  simulation e a bootstrap)

```python
forecasts = res.forecast(horizon=5, start=split_date)
forecasts.variance[split_date:].plot()
```

```python
forecasts.mean #tem tambm o variance e residual_variance
#h1  uma observao a frente, h5 so 5 a frente, etc

#a mdia  assumida constante
#ele s d o forecast at onde vai o dataset original
```

```python
#valores que no esto no forecast viram nan
day = dt.timedelta(1)
print(forecasts.variance[split_date - 5 * day:split_date + 5 * day])
```

<h1> GJR-GARCH </h1>
Inclui one lag of an asymmetric shock, onde a varincia fica:
$$\sigma^{2}_{t} = \omega + \alpha \epsilon^{2}_{t-1}  + \gamma \epsilon^{2}_{t-1} I_{[\epsilon_{t-1<0}]} + \beta \sigma^{2}_{t-1}$$ onde $I$ fica 1 se h um lag.

```python
am = arch_model(returns, p=1, o=1, q=1) #o 'o'  o lag. no garch,  zero
res = am.fit(update_freq=5)
res
```

```python
#separando at essa data e o resto ele faz o forecast
split_date = dt.datetime(2010,1,1) 
res = am.fit(last_obs=split_date, disp='off')

forecasts = res.forecast(horizon=5, start=split_date)
forecasts.variance[split_date:].plot()
```

```python
am = arch_model(returns, p=1, o=1, q=1, mean='ARX') #o 'o'  o lag. no garch,  zero
#res = am.fit(update_freq=5)
#res
```

```python
#separando at essa data e o resto ele faz o forecast
split_date = dt.datetime(2010,1,1) 
res = am.fit(last_obs=split_date, disp='off')

forecasts = res.forecast(horizon=5, start=split_date)
forecasts.variance[split_date:].plot()
```
