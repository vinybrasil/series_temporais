```python
import pandas_datareader.data as web
import datetime as dt
import arch.data.sp500
from arch import arch_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#st = dt.datetime(1988,1,1)
#en = dt.datetime(2019,1,1)
#data = web.get_data_yahoo('^GSPC', start=st, end=en)
data = arch.data.sp500.load()
```

```python
market = data["Close"]
market_train = market #[:'2017-12-29']
market_train['2017-12-29':] = np.nan
```

```python
data
```

```python
#data.to_csv("sp500.csv")
returns = 100 * market_train.pct_change().dropna()
```

```python
am = arch_model(returns)
res = am.fit()
```

```python
forecasted_variance = res.model.forecast(res.params, horizon = 252, start='2017-12-29',  align='target')._variance
```

```python
forecasted_variance
```

```python
plt.plot(forecasted_variance.loc['2018-12-31'].values)

#roda 
plt.plot(data["Close"]['2017-12-29':].pct_change().values)
```

```python
forecasted_variance = res.model.forecast(res.params, horizon = 252,  align='origin')._variance
```

```python
forecasted_variance
```

```python
plt.plot(forecasted_variance.loc['2018-12-31'].values)

#roda 
plt.plot(data["Close"]['2017-12-29':].pct_change().values)
```
