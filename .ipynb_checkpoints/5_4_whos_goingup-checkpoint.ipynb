```python
import tensorflow as tf
print(tf.__version__)
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([[1,2],
              [4,3],
              [5,6],
              [2,3]])

#x = x.reshape(4, 2, 1)
```

```python
x.shape
```

```python
y = np.array([[0,1],
              [1,0],
              [0,1],
              [0,1]])
```

```python
model = Sequential()
#model.add(LSTM(1))
model.add(Dense(2))# input_shape=(2,1)))
model.add(Dense(8, kernel_initializer='uniform', activation='softmax'))
model.add(Dense(2))
#model.add(Dense(1))

opt = tf.keras.optimizers.Adam(learning_rate=0.01)

model.compile(loss='mae', optimizer=opt)
history = model.fit(x, np.array(y), epochs = 300, batch_size=16, verbose=2)
plt.plot(history.history['loss'])
```

```python
model.predict(np.array([[1,2]]))
```

```python
x = np.array([[1,2, 15, 46, 101 ],
              [4,3, -2, -6, -45],
              [5,6, 42, 49, 51],
              [2,1, 0, -1, -2]])

y = np.array([[0,1],
              [1,0],
              [0,1],
              [1, 0]])
```

```python
model = Sequential()
#model.add(LSTM(1))
model.add(Dense(5))# input_shape=(2,1)))
model.add(Dense(8, kernel_initializer='uniform', activation='softmax'))
model.add(Dense(2))
#model.add(Dense(1))

opt = tf.keras.optimizers.Adam(learning_rate=0.01)

model.compile(loss='mae', optimizer=opt)
history = model.fit(x, np.array(y), epochs = 300, batch_size=16, verbose=2)
plt.plot(history.history['loss'])
```
