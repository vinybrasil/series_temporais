```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sklearn.metrics as skm
def show_errors(ytrue, ypred):
    print("MAE: ", skm.mean_absolute_error(ytrue, ypred))
    print("MSE: ", skm.mean_squared_error(ytrue, ypred))
    print("Median AE: ", skm.median_absolute_error(ytrue, ypred))
```

```python
baranya['log_price'] = np.log(baranya['cases'])

baranya['pct_change'] = baranya['log_price'].diff()
baranya['pct_change'][0] = 0

baranya
```

```python
baranya['stdev'] = baranya['pct_change'][0:].rolling(window=2, center=False).std()
baranya['Volatility'] = baranya['stdev'] * (58**0.5) # Annualize.
```

```python
baranya
```

```python
plt.plot(baranya['Volatility'])
```

```python
baranya_2 = baranya.dropna()
vol = baranya_2["Volatility"] * 100
```

```python
from arch import arch_model
#am = arch_model(vol[0:58*8], vol='Garch', p=1, o=0, q=1, dist='Normal')
am = arch_model(vol, vol='Garch', p=1, o=0, q=1, dist='Normal')

res = am.fit(disp='off')
display(res.summary())
```

```python
res.conditional_volatility
```

```python
baranya_2['forecast_vol'] = 0.01 *  np.sqrt(res.params['omega'] + res.params['alpha[1]'] * res.resid**2 + 
                                   res.conditional_volatility**2 * res.params['beta[1]'])
```

```python
show_errors(baranya_2["Volatility"].values, baranya_2["forecast_vol"].values)
```

```python
plt.figure(figsize=(16,6))
baranya_2["forecast_vol"].plot()

baranya_2["Volatility"].plot()
plt.title("Real Rolling Volatility vs Forecast by GARCH(1,1)")
plt.legend()
plt.show()
```

```python
am = arch_model(vol, mean='HARX', vol='Garch', p=1, o=0, q=1, dist='t')

res = am.fit(disp='off')
display(res.summary())
```
