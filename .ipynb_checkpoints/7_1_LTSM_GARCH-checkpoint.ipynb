```python
#source: https://github.com/Bturan19/Neural-Garch-Hybrid-Model-Implementation/blob/master/Final_Proje_Denemesi.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# plotly is an advanced visualization tool
import plotly

```

```python
data = pd.read_csv('datasets/hungary_chickenpox.csv')
```

```python
budapest = data[['Date', 'BUDAPEST']]
```

```python
budapest
```

```python
plt.plot(budapest.BUDAPEST)
```

```python
budapest['log_price'] = np.log(budapest['BUDAPEST'])

budapest['pct_change'] = budapest['log_price'].diff()

budapest['stdev'] = budapest['pct_change'].rolling(window=30, center=False).std()
budapest['Volatility'] = budapest['stdev'] * (252**0.5) # Annualize.
```

```python

plt.figure(figsize=(16,6))
budapest['Volatility'].plot()
plt.title("Rolling Volatility With 30 Time Periods By Annualized Standard Deviation")
plt.show()

```

```python
budapest = budapest.dropna()
vol = budapest["Volatility"] * 100
```

```python
#GARCH

from arch import arch_model
am = arch_model(vol, vol='Garch', p=1, o=0, q=1, dist='Normal')

res = am.fit(disp='off')
display(res.summary())
```

```python
budapest['forecast_vol'] = 0.1 * np.sqrt(res.params['omega'] + res.params['alpha[1]'] * res.resid**2 + 
                                   res.conditional_volatility**2 * res.params['beta[1]'])
```

```python
display(budapest.tail(10))
```

```python
plt.figure(figsize=(16,6))
budapest["Volatility"].plot()
budapest["forecast_vol"].plot()
plt.title("Real Rolling Volatility vs Forecast by GARCH(1,1)")
plt.legend()
plt.show()

```

```python
def rmse_tr(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
skor = rmse_tr(budapest.loc[budapest.index[1000:], 'forecast_vol'], budapest.loc[budapest.index[1000:], 'Volatility'])
print("Root Mean Squared Error of the model is calculated as ",skor)
```

```python
budapest.shape
```

```python
training_set = budapest.iloc[:, -2:].values
# 100 timestep ve 1 Ã§Ä±ktÄ± ile data yapÄ±sÄ± oluÅŸturalÄ±m
X_train = []
y_train = []
for i in range(1, budapest.shape[0]):
    X_train.append(training_set[i-1:i, 0])
    y_train.append(training_set[i,0])
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
budapest
```

```python


# Importing the Keras libraries and packages
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dropout

```

```python
# Initialising the RNN
regressor = Sequential()

# Adding the first LSTM layer and some Dropout regularisation
regressor.add(LSTM(units = 10, return_sequences = True, input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(0.1))

# Adding a second LSTM layer and some Dropout regularisation
regressor.add(LSTM(units = 10, return_sequences = True))
regressor.add(Dropout(0.1))

# Adding a third LSTM layer and some Dropout regularisation
regressor.add(LSTM(units = 10, return_sequences = True))
regressor.add(Dropout(0.1))

# Adding a fourth LSTM layer and some Dropout regularisation
regressor.add(LSTM(units = 10))
regressor.add(Dropout(0.1))

# Adding the output layer
regressor.add(Dense(units = 1))

# Compiling the RNN
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

# Fitting the RNN to the Training set
regressor.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

```python
regressor.save('my_modelp1.h5')
```

```python


from tensorflow.keras.models import load_model
regressor = load_model('my_modelp1.h5')

```

```python
predicted_stock_price = regressor.predict(X_train)
```

```python
# Visualising the results
plt.figure(figsize=(18,6))
plt.plot(budapest.iloc[1:, -2:].values, color = 'red', label = 'Observed Volatility')
plt.plot(predicted_stock_price, color = 'blue', label = 'Predicted Volatility By LSTM')
plt.title('Real Rolling Volatility vs Forecast of LSTM')
plt.xlabel('Time')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```

```python


skor2 = rmse_tr(predicted_stock_price, np.array(budapest.loc[budapest.index[1:], 'Volatility']))
print("Root Mean Squared Error of the model is calculated as ",skor2)

```

```python
training_set = budapest.iloc[:, -2:].values
# 100 timestep ve 1 Ã§Ä±ktÄ± ile data yapÄ±sÄ± oluÅŸturalÄ±m
X_train = []
y_train = []
for i in range(1, budapest.shape[0]-1):
    X_train.append(training_set[i-1:i,:])
    y_train.append(training_set[i,0])
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 2))
```

```python
# Initialising the RNN
regressor = Sequential()

# Adding the first LSTM layer and some Dropout regularisation
regressor.add(LSTM(units = 10, return_sequences = True, input_shape = (X_train.shape[1], 2)))
regressor.add(Dropout(0.1))

# Adding a second LSTM layer and some Dropout regularisation
regressor.add(LSTM(units = 10, return_sequences = True))
regressor.add(Dropout(0.1))

# Adding a third LSTM layer and some Dropout regularisation
regressor.add(LSTM(units = 10, return_sequences = True))
regressor.add(Dropout(0.1))

# Adding a fourth LSTM layer and some Dropout regularisation
regressor.add(LSTM(units = 10))
regressor.add(Dropout(0.1))

# Adding the output layer
regressor.add(Dense(units = 1))

# Compiling the RNN
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

# Fitting the RNN to the Training set
regressor.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

```python
regressor.save('my_modelp2.h5')
```

```python
from tensorflow.keras.models import load_model
regressor = load_model('my_modelp2.h5')
```
